```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('minutes_and_certificates.csv')
```

```python
df = data.copy()
```

```python
df.head()
```

```python
X = df[['minutes_watched']]
y = df['certificates_issued']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=365)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.intercept_, model.coef_
```

```python
y_pred = model.predict(X_test)
r_squared = model.score(X_train, y_train)
print(f"R-squared: {r_squared}")
```

```python

predicted_certificates = model.predict([[1200]])
predicted_certificates_rounded = round(predicted_certificates[0])
print(f"Predicted certificates for 1200 minutes watched: {predicted_certificates_rounded}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.title('Actual vs Predicted Certificates Issued')
plt.xlabel('Actual Certificates Issued')
plt.ylabel('Predicted Certificates Issued')
plt.show()
```
